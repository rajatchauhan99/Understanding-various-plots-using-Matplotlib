```python
# we can change the figure size

fig, axes = plt.subplots(figsize=(16,9), dpi = 300)  # pixels per inch 300 number for a very good quality

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(50,4), index = pd.date_range('today', periods = 50), columns = ["A","B","C","D"])

df = df.cumsum()
axes.plot(df)
```

```python
# we can set title and other stuff

fig, axes = plt.subplots(figsize = (10,5), dpi = 300)

axes.plot(df)

axes.set_title("My Plot")
axes.set_xlabel("x-axis")
axes.set_ylabel("y-axis")

axes.legend(["A","B","C","D"])
```

```python
# setting properties such as line color, transparency and more

fig, axes = plt.subplots(dpi = 300)

x = np.linspace(0,10,25)

axes.plot(x, x+1, color = 'green', alpha = 0.1)   # alpha makes it transparent (setting opacity)
axes.plot(x, x+2, color = 'grey')
axes.plot(x, x+3, color = 'orange')


# 0.1 alpha makes it very blurry so 0.5 is a good practise
```

```python
# In same way, we can play with line width as well
```

```python
fig, ax = plt.subplots(dpi = 300)

ax.plot(x, x+1, color = 'blue', linewidth = 0.25)
ax.plot(x, x+2, color = 'blue', linewidth = 1)
ax.plot(x, x+3, color = 'blue', lw = 3)
```

```python
# we can use different line styles

fig, ax = plt.subplots(dpi = 300)

ax.plot(x, x + 1 , lw = 1.5, linestyle = '--')

ax.plot(x, x + 2 , lw = 1, linestyle = '-.')

ax.plot(x, x + 3 , lw = 2, linestyle = ':')
```

```python
# marker

fig, ax = plt.subplots(dpi = 300)

ax.plot(x, x+1, lw = 1, marker = 'o')   # this gives dot markers for the 25 data points which is the most common


ax.plot(x, x+2, lw = 2, marker = '+')


ax.plot(x, x+3, lw = 2, marker = 's')  # s for square


ax.plot(x, x+4, lw = 2, marker = '1')    # kind of hatch marks


ax.plot(x, x+6, lw = 2, marker = 's', markersize = 5, markerfacecolor = 'green')
```

```python
# Draw a radar chart (for that we choose polar = True)

fig = plt.figure(figsize = (6,6), dpi = 200)

ax = fig.add_axes([0.0, 0.0, 0.6, 0.6], polar = True)  # setting the axes for our plot (left, bottom, width, height)

t = np.linspace(0, 2*np.pi, 100)       # getting 100 equally spaced numbered between 0 and 2*pi

ax.plot(t,t, lw = 3)
```

```python
# Draw Counter Maps

from matplotlib import cm as cm
from matplotlib import pyplot as plt
import numpy as np

delta = 0.025

x = np.arange(-3,3, delta)
y = np.arange(-2,2, delta)
X,Y = np.meshgrid(x,y)
X,Y
Z1 = np.exp(-X**2-Y**2)
Z2 = np.exp(-(X-1)**2-(Y-1)**2)
Z = (Z1-Z2)*2

print(X)

print(Y)
```

```python
fig, ax = plt.subplots(dpi = 100)

CS = ax.contour(X,Y,Z)   # X and Y are coordinates and Z is the actual height

ax.clabel(CS, inline = 1, fontsize = 10)

ax.set_title("Contour map")
```

```python
# here 0 line shows the sea level, 
# other numbers shows the highs and lows of mountains pits for negative sides

# in data science this can be used to represent 
# two different values such as profit and loss
```

Write a python program to create a pie chart of the popularity of programming languages

```python
languages = ["Python", "Java", "PHP", "JavaScript", "C#", "C++"]

popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

colors = ['red', 'green', 'blue', 'yellow', 'silver', 'pink']

explode = (0.1,0,0,0,0, 0)

plt.pie(popularity, colors = colors, labels = languages, explode = explode, 
       autopct = '%1.1f%%', shadow = True, startangle = 90)


# autopct puts percentage values on the slicers

# shadow gives a 3D feel
```
